### 3. Subplots (With Seaborn) - 3 pts
3.1 Find Top 10 neighbourhoods with the **most number** of AirBnB units available and show them in an horizontal bar chart. (1 pt)

Any solution is accepted as long as it follows the mentioned points:
- Find Top 10 most frequent neighbourhoods
- Horizontal Bar Chart displays the "Units available per neighbourhood" of these Top 10 most frequent neighbourhoods
- The x-axis is Units Available
- The y-axis is Neighbourhoods

```python
# YOUR CODE HERE
top_10_hood = listings_df['neighbourhood'].value_counts().index[:10]
#top_10_hood
units_ava = listings_df['neighbourhood'].value_counts().values[:10]
#units_ava

bar_plot=sns.barplot(x= units_ava, y=top_10_hood)

plt.title('Units available per neighbourhood')
plt.xlabel('Units Available')
plt.ylabel('Neighbourhoods')
```

3.2 Find the mean price, mean number_of_reviews, mean availability_365 for neighbourhoods found in part **3.1** and create a lineplot for the mean values of price, number_of_reviews and availability_365. (1 pt)

Any solution is accepted as long as it follows the mentioned points:
- The Line Plot displays "Mean values per neighbourhood" of price, number_of_reviews, and availability_365
- The x-axis is Neighbourhood
- The y-axis is Mean Values

```python
# YOUR CODE HERE
#passengers[passengers['Fare'].isin(top10_fares2)]

listings_df_filtered = listings_df[listings_df['neighbourhood'].isin(top_10_hood)]
#listings_df_filtered  
mean_values = listings_df_filtered.groupby('neighbourhood')[['price', 'number_of_reviews', 'availability_365']].agg('mean')
#mean_values

line_plot = sns.lineplot(data=mean_values)
#line_plot.set_xticklabels(line_plot.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.xticks(rotation=45,horizontalalignment='right')
plt.title('Mean values per neighbourhood')
plt.xlabel('Neighbourhood')
plt.ylabel('Mean Values')
```

3.3 Create a sublot of size 1x2 and add both the plots from part **3.1** and **3.2** to this figure. (1 pt)

Any solution is accepted as long as it follows the mentioned points:
- A subplot of size 1x2 grid layout (refers to 1 row x 2 columns)

```python
# YOUR CODE HERE
#sns.set()

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

sns.barplot(x= units_ava, y=top_10_hood, ax=axes[0])
axes[0].set_title('Units available per neighbourhood')
axes[0].set_xlabel('Units Available')
axes[0].set_ylabel('Neighbourhoods')

sns.lineplot(data=mean_values, ax=axes[1])
axes[1].tick_params(axis='x', rotation=90)
axes[1].set_title('Mean values per neighbourhood')
axes[1].set_xlabel('Neighbourhood')
axes[1].set_ylabel('Mean Values')
```

### 4. Correlation Heatmap using Seaborn - 1 pt

### 5. Draw a combination of boxplot and kernel density estimate (aka Violin Plot) for Price vs Neighbourhood - 1 pt
Use 10 most popular neighbourhoods found in 3.1 with additional filter of removing all the listings which are more than 1500$ in price.

Any solution is accepted as long as it follows the mentioned points:
- The x-axis is Neighbourhood
- The y-axis is Price
- Listings which are less or equal than 1500 in price

```python
# YOUR CODE HERE
listings_df_filtered_1500 = listings_df_filtered[listings_df_filtered['price']<=1500]
#listings_df_filtered_1500[['neighbourhood', 'price']]
sns.set()

fig, axes = plt.subplots(1,2, figsize=(12, 6))                                                                    
                         
sns.boxplot(data=listings_df_filtered_1500, x='neighbourhood', y='price', ax=axes[0])
axes[0].tick_params(axis='x', rotation=90)
axes[0].set_title('Boxplot')

sns.violinplot(data=listings_df_filtered_1500, x='neighbourhood', y='price', ax=axes[1])
axes[1].tick_params(axis='x', rotation=90)
axes[1].set_title('Violinplot')
```
